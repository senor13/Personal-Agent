```python
prompt = """ 
    Your job is to extract important information from the conversation history which is given. Only extract clear and specific
    facts about the user which is worth remembering long term
    Ignore small talk and questions. Return a JSON list like:
    [{"category":"about user","content":"User's name is Sagrika"}]
    Return only raw JSON, no markdown, no code fences

    """
```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv() # reads .env file and loads variables into os.environ


client = OpenAI(api_key = os.environ["OPENAI_API_KEY"])
```

```python
import json
conversation_history = [{"role":"user","content": "user's name is Sagrika"}]
message = conversation_history + [{"role":"user","content":prompt}]
response = client.chat.completions.create(model = "gpt-4o-mini", messages = message, tools = None)

   
```

```python
response.choices[0].message.content
```

```python
facts = json.loads(response.choices[0].message.content)
```

```python
import os
from googleapiclient.discovery import build
youtube = build("youtube","v3",developerKey = os.environ["YOUTUBE_API_KEY"])
```

```python
search_list = youtube.search().list(q='latest videos on AI agents', type="video", part="snippet", maxResults=5).execute()
```

```python
search_list
```

```python
search_list["items"][0]["snippet"]["title"]
```

```python
video_ids = []
for item in search_list["items"]:
    video_ids.append(item["id"]["videoId"])
search_list_2 =  youtube.videos().list(part="statistics", id=",".join(video_ids)).execute()
```
